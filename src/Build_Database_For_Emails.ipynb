```python
import os
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.lancaster import LancasterStemmer
import email
stop_words = set(stopwords.words('english'))
st = LancasterStemmer()
```

```python
def check_file(file_path):#input a path and return all the subfiles in it
    os.chdir(file_path)
    all_file = os.listdir()
    files = []
    for f in all_file:# find files recursively
        if os.path.isdir(f):
            files.extend(check_file(file_path+'/'+f))
            os.chdir(file_path)
        else:
            files.append(os.path.abspath(f))
    return files
```

```python
tokenizer = RegexpTokenizer(r'[A-Za-z][A-Za-z][A-Za-z]+')
from nltk.stem.snowball import EnglishStemmer
from nltk.stem.porter import PorterStemmer
es=EnglishStemmer()
ps=PorterStemmer()
keyworddict={}
for i in range(0, len(file_list)):
    files = open(file_list[i], 'r')
    tokendict={}
    tempEmail = email.message_from_string(files.read())
    tempMessage = tempEmail.get_payload()
    tokens=[]
    tokens+=(tokenizer.tokenize(str(tempEmail['Subject'].lower())+str(tempMessage.lower())))
#     print (tokens)
    tokens_withoout_Stopword = [w for w in tokens if not w in stop_words]
#     print (tokens_withoout_Stopword)
    tokens_stemmed = [es.stem(word) for word in tokens_withoout_Stopword]
#     print(tokens_stemmed)
    for token in tokens_stemmed:
        if token in tokendict:
            tokendict[token] += 1
        else:
            tokendict[token] = 1
    keyworddict[file_list[i]]=tokendict
    files.close()
print(keyworddict)
```

```python
inverteddict={}
for i in range(len(keyworddict)):
    path=list(keyworddict.keys())[i]
    dicts=list(keyworddict.values())[i]
    for j in range(len(dicts)):
        words=list(dicts.keys())[j]
        times=list(dicts.values())[j]
        if words not in inverteddict:
            inverteddict[words]={}
        inverteddict[words][path]=times
print(inverteddict)      
```
